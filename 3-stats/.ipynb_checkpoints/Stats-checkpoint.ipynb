```python
import matplotlib.pyplot as plt
from scipy.stats import skew

fig, ax = plt.subplots(figsize=(8, 6))
mu = 0.3858 # mean of distribution in %
sigma = 2.7622  # standard deviation of distribution  in %

x = df_T10['Return']*100

num_bins = 50

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

ax.plot(bins, y, )
ax.set_xlabel('Daily Return %')
ax.set_ylabel('Probability density')
ax.grid()
ax.set_title(r'Histogram of Treasury Yield 10 Years: $\mu$=0.3858%, $\sigma$=2.7622%')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()

plt.show()
```

# 0700.HK

```python
df_700 = yf.download(
    "0005.HK", 
    start="2025-01-01", 
    end="2025-12-02",
    auto_adjust=True
).droplevel('Ticker', axis=1)

# 計日回報率
df_700['Return'] = df_700["Close"].pct_change(1)

# 去掉 NaN
df_700 = df_700.dropna()

# 正確 Z-score 計法
df_700['Zscore'] = (df_700['Return'] - df_700['Return'].mean()) / df_700['Return'].std()

df_700['Close_Zscore'] = (df_700['Return'] - df_700['Return'].mean()) / df_700['Return'].std()
```

```python
fig, ax = plt.subplots(
    nrows=3,
    ncols=1,
    sharex=True,
    figsize=(12,6),
    gridspec_kw={"height_ratios":[2,1,1]}  # <-- 係 height_ratios !!
)

ax[0].plot(df_700.index, df_700.Close, label='Close') # 第1個圖
ax[0].grid(True)
ax[0].legend(loc='upper right')
ax[0].set_ylabel('Close Price')
ax[0].set_title('')
ax[1].plot(df_700.index, df_700.Zscore, 'green', label = 'Z-Score')  # 第2個圖
ax[1].grid(True)
ax[1].legend(loc='upper right')
ax[1].set_ylabel('Z-Score')
ax[2].plot(df_700.index, df_700.Zscore, 'orange', label = 'Close Z-Score') # 第3個圖
ax[2].grid(True)
ax[2].legend(loc='upper right')
ax[2].set_ylabel('Close Z-Score') #Z score 計出嚟數值，比較，通常within -4 ＋4, 計比例
plt.tight_layout()
plt.show()
```

```python
kurtosis(df_700['Return'], fisher=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew

fig, ax = plt.subplots(figsize=(8, 6))

mu =  df_700['Return'].mean()*100  # mean of distribution
sigma = df_700['Return'].std()*100  # standard deviation of distribution

x = df_700['Return']*100

num_bins = 50

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

ax.plot(bins, y )

ax.set_xlabel('Daily Return %')
ax.set_ylabel('Probability density')
ax.grid()
ax.set_title(fr'Histogram of 0700.HK: $\mu$={mu:.2f}%, $\sigma$={sigma:.2f}%')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()

plt.show()
```

# Hypothesis Testing
- Calculate the T-test for the mean of ONE group of scores.
- This is a test for the null hypothesis that the expected value (mean) of a sample of independent observations a is equal to the given population mean, popmean.

```python
import numpy as np
import scipy.stats as stats

data = [15.9,16.1,16.3,16.25,16.3,16.22,16.2,16.1,16.15,16.08,16.32,16.3]

print(f"sample size: {len(data)}, sample mean: {sum(data)/len(data)}, standard deviation: {np.std(data,axis=0)}")
#perform one sample t-test
stats.ttest_1samp(a=data, popmean=16.1)
```
