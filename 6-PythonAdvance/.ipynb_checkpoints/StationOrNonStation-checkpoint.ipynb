# Test a period for Stationarity or Non-Stationary
https://www.statsmodels.org/stable/examples/notebooks/generated/stationarity_detrending_adf_kpss.html
- [**`Augmented Dickey-Fuller Test`**](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test):
ADF tests the null hypothesis that a time series is non-stationary.<BR>
If the p-value is low (e.g., < 0.05), you reject the null hypothesis and conclude the series is <u>**stationary**</u>.
<p></p>
- [**`KPSS Test`**](https://en.wikipedia.org/wiki/KPSS_test):
KPSS tests the null hypothesis that a time series is stationary.<BR>
If the p-value is low (e.g., < 0.05), you reject the null hypothesis and conclude the series is <u>**non-stationary**</u>.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning
warnings.simplefilter('ignore', InterpolationWarning)
```

# Test 1 - ADF

```python
# Download stock price data
prices = yf.download("AAPL", start="2025-01-01", end="2025-03-01", 
                     progress=False, auto_adjust=True).droplevel('Ticker',axis=1)['Close']

# Augmented Dickey-Fuller Test
adf_result = adfuller(prices, autolag="AIC")
print("ADF Test Statistic:", adf_result[0])
print("ADF P-Value:", adf_result[1])
if adf_result[1] < 0.05:
    print("The series is stationary based on the ADF test (p-value < 0.05).")
else:
    print("The series is non-stationary based on the ADF test (p-value >= 0.05).")
print()

# KPSS Test
from statsmodels.tsa.stattools import kpss
kpss_result = kpss(prices, regression='c')
print("\nKPSS Test Statistic:", kpss_result[0])
print("KPSS P-Value:", kpss_result[1])
if kpss_result[1] < 0.05:
    print("The series is non-stationary based on the KPSS test (p-value < 0.05).")
else:
    print("The series is stationary based on the KPSS test (p-value >= 0.05).")
print()

prices.plot()
```

```python
def adf_test(timeseries):
    print("Results of Dickey-Fuller Test:")
    dftest = adfuller(timeseries, autolag="AIC")
    dfoutput = pd.Series(
        dftest[0:4],
        index=[
            "Test Statistic",
            "p-value",
            "#Lags Used",
            "Number of Observations Used",
        ],
    )
    for key, value in dftest[4].items():
        dfoutput["Critical Value (%s)" % key] = value
    print(dfoutput)
```

```python
x = np.linspace(0, 10, 1000)
y = 4 + 1 * np.sin(2 * x)
plt.figure(figsize=(5,2))
plt.plot(y)
plt.show()
```

```python
adf_test(y)
```

# If the KPSS test suggests stationarity and the ADF test suggests non-stationarity, the series might be stationary around a trend.
- The ADF test has lower power to detect stationarity, meaning it can fail to reject the null hypothesis even if the series is stationary.
- The KPSS test is overly sensitive to small deviations from stationarity, meaning it can falsely reject the null hypothesis of stationarity.

# Test 2 - KPSS

```python
# Download stock price data
prices2 = yf.download("AAPL", start="2025-01-01", end="2025-03-01", 
                     progress=False, auto_adjust=True).droplevel('Ticker',axis=1)['Close']

# KPSS Test
from statsmodels.tsa.stattools import kpss
kpss_result = kpss(prices2, regression='c')
print("\nKPSS Test Statistic:", kpss_result[0])
print("KPSS P-Value:", kpss_result[1])
if kpss_result[1] < 0.05:
    print("The series is non-stationary based on the KPSS test (p-value < 0.05).")
else:
    print("The series is stationary based on the KPSS test (p-value >= 0.05).")
print()

prices2.plot()
```

```python
def kpss_test(timeseries):
    print("Results of KPSS Test:")
    kpsstest = kpss(timeseries, regression="c", nlags="auto")
    kpss_output = pd.Series(
        kpsstest[0:3], index=["Test Statistic", "p-value", "Lags Used"]
    )
    for key, value in kpsstest[3].items():
        kpss_output["Critical Value (%s)" % key] = value
    print(kpss_output)
```

```python
x = np.linspace(0, 10, 1000)
y = 4 + 1 * np.sin(2 * x)
plt.figure(figsize=(5,2))
plt.plot(y)
plt.show()
```

```python
kpss_test(y)
```

# 2. Test non-stationary

```python
plt.plot(x+y)
plt.figure(figsize=(5,2))
plt.show()
```

```python
adf_test(x+y)  # p-value >= 0.05, this is non-stationary
```

```python
kpss_test(x+y) # p-value < 0.05 , this is non-stationary
```

```python
kpss_test(x+y) # p-value < 0.05 non-stationary
```

# 3. Test stationary plus trend

```python
plt.plot(np.concatenate([y, (x+y)[:100]]))
plt.figure(figsize=(5,2))
plt.show()
```

```python
adf_test(np.concatenate([y, (x+y)[:100]]))  # p-value >= 0.05, this is non-stationary
```

```python
kpss_test(np.concatenate([y, (x+y)[:100]]))   # p-value > 0.05 stationary
```
