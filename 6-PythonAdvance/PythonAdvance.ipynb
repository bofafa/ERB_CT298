# Generator

```python
def yield_hello():
    yield "Hello1"
    yield "Hello2"
    yield "Hello3"

yield_hello()
```

```python
def genFibonnaci(max_num):
    a, b = 0,1
    count = 0
    while count < max_num:
        yield a
        c = a + b
        a = b
        b= c
        count +=1

fib = list(i for i in genFibonnaci(20))

print(fib)
```

```python
ratio = []
for i in range (len(fib)-1):
    ratio.append(fib[i]/fib[i+1])
print('Golden Ration:', sum (ratio)/len(ratio) )
```

# cache

```python
from functools import lru_cache
import time

start_time = time.time()

@lru_cache(maxsize = None)
def fibonacci(n):
    if n < 2:
        return n
    else:
        return fibonacci(n-1) + fibonacci (n-2)
fib = [fibonacci(i) for i in range(200)]

end_time = time.time()

print(fib[-1])
print(f"It took {end_time  - start_time:.2f} seconds to execute.")
print (fibonacci.cache_info())
```

# words

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import math

# Load the text file
filename = "news.txt"
try:
    with open(filename, encoding='utf-8') as f:
        contents = f.read()
except FileNotFoundError:
    print(f"Sorry, the file {filename} does not exist.")
except Exception as e:
    print(f"Error: {e}")

# Sentences and words analysis
sentences = nltk.sent_tokenize(contents) #拆句
words = nltk.word_tokenize(contents) #拆字
num_words = len(words)  #拆字數
num_sentences = len(sentences)  #拆句數
word_per_sen = round((num_words / num_sentences), 2) # <- round 2 digi
read_minute = math.ceil(num_words/250) #use one min, 250字 1分鐘，字數/250， ceiling 拎最大數

## Sentiment analysis
# Load positive words bank
pos_words = open("positive-words.txt", 'r', encoding="ISO-8859-1").read().split()

# Load negative words bank
neg_words = open("negative-words.txt", 'r', encoding="ISO-8859-1").read().split() 

num_pos,num_neg = 0,0 #設定 0
pos_list,neg_list = [],[] #開LIST

for w in words:  #bollian
    if w in pos_words:   # w = word 
        pos_list.append(w) # pos 入LIST
        num_pos += 1 
    if w in neg_words:
        neg_list.append(w) # neg 入LIST 
        num_neg += 1 
        
sentiment_ratio = round( (num_pos / (num_pos + num_neg)), 2)
if sentiment_ratio > 0.5:
    sentiment = 'positive'
else:    
    sentiment = 'negative'

# readability
if word_per_sen <= 15:
    readability = 'very comprehensible'
elif word_per_sen > 40:   
    readability = 'hard to comprehend'
else:    
    readability = 'comprehensible'

# Output Analysis 
print(f"The content is {readability}. The content is likely to be {sentiment} with ratio {sentiment_ratio} in positive words.")
print(f"Reading time is approximately {read_minute} mins.")
print(f"The file {filename} has about {num_sentences} sentences, {num_words} words, average words per sentence is {word_per_sen}.")
print(f"Total positive words {num_pos} : {pos_list}")
print(f"Total negative words {num_neg} : {neg_list}")
```
