# Slicing 1-D array

```python
arr2 = np.array([1,2,3])
print(arr2[0]) #1
print(arr2[1]) #2
print(arr2[-1])#3
print(arr2[:]) #[1 2 3]
print(arr2[1:]) #[ 2 3]
print(arr2[:2]) #[1 2]
```

```python
arr_2D = np.array([[1,2],[3,4],[5,6],[7,8]])
print (arr_2D)
arr_2D.shape  #(4,2) =  4 row, 2 colnum 
```
