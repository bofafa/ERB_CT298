# .delete

```python
arr = np.arange(12).reshape(3,4)
arr
```

```python
np.delete(arr, 2, axis=0) #<=v(arr, 第幾行，axis = 0 /1 row/ colnum)
```

```python
np.delete(arr, np.s_[2], axis=1) #<-- handle 多D data 會run 快D
```

```python
np.delete(arr, 2, axis=1) #<- 同 np.s_[2] 一樣
```

# .unique #<- 搵array label

```python
b = np.array([1,2,1,2,5,3,3])
np.unique(b)
```

```python
r = np.array([[1,0,0],[1,0,0],[2,4,4]])
r
```

```python
r = np.array([[1,0,0],[1,0,0],[2,4,4]])
np.unique(r, axis = 0) #<- return unique row
```

```python
a = np.array ([1,2,6,4,2,3,2])
values, counts = np.unique(a, return_counts = True)
values # arr of unique val
```
