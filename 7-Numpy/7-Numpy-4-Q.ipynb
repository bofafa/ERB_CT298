## 28. Find the position of missing value in numpy array
<pre>
Find: 
1. Number of missing values
2. Row number that contains missing values
3. The NaN row value

Given Input:
arr = np.array([[1, 2, 3],
                [4, np.NaN, 6],
                [7, 8, 9],
                [np.NaN, np.NaN, np.NaN]
               ])

Desired Output:
Number of missing values: 
 4
Row contains missing values: 
 (array([1, 3]),)
The NaN row: 
 [[ 4. nan  6.]
 [nan nan nan]]</pre>

```python
arr = np.array([[1,2,3],
                [4,np.nan,6],
                [7,8,9],
                [np.nan,np.nan,np.nan]
               ])
arr
```

```python
print("Number of missing values: \n",)
print("Row contains missing values: \n",  )
print("The NaN row: \n",)
```

```python
# 1. Number of missing values
num_missing = np.isnan(arr).sum()
print("Number of missing values:\n", num_missing)

# 2. Row number that contains missing values
rows_with_nan = np.where(np.isnan(arr).any(axis=1))
print("Row contains missing values:\n", rows_with_nan)

# 3. The NaN row value
nan_rows = arr[rows_with_nan]
print("The NaN row:\n", nan_rows)
```

array([[ 2.,  4.,  6.],
