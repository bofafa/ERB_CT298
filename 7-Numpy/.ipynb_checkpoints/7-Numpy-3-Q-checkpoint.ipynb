## 14. Extract all numbers between a given range from a numpy array
<pre>Get all items between 5 and 10 from a.

Given Input:
a = np.array([2, 6, 1, 9, 10, 3, 27])

Desired Output:
array([6, 9, 10]) 
or
[6, 9, 10]
</pre>

```python
a = np.array([2, 6, 1, 9, 10, 3, 27])

# Method 1
```

```python
# Method 2
# [ expression for item in list if conditional ]
arr = a[()]
```

```python
# Method 3
# use lambda
```

```python
arr = a[(a >= 5) & (a <= 10)]
arr
```

```python
arr = a[(a >= 5) & (a <= 10)].tolist()
arr
```

```python
arr = list(map(int, filter(lambda x: 5 <= x <= 10, a)))
arr
```

## 15. Make a python function that handles scalars to work on numpy arrays
<pre>Convert the function maxx that works on two scalars, to work on two arrays.

Given Input:
def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

maxx(1, 5)
$> 5

Desired Output:
a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
pair_max(a, b)
$> array([6, 7, 9, 8, 9, 7, 5])
or
[6, 7, 9, 8, 9, 7, 5]
</pre>

```python
a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])

def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y


# Method 1  (np.vectorize)

pair_max = np.vectorize(maxx)  #vectorize 用嚟比較 2D array, maxx 只做到1 d
pair_max(a,b)
```

```python
# Method 2 (np.frompyfunc)
pair_max = np.frompyfunc(maxx, 2, 1)
pair_max(a,b)
```

```python
# Method 3 (Lambda)
list(map(lambda x,y: maxx(x,y), a,b))
```
