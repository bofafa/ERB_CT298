# np.ndenumerate 多維陣列
Multidimensional index iterator.
Return an iteratory yielding pairs of array corrdinates and values

✔ 同時取到：

每個元素嘅 座標（index）

每個元素嘅 值（value）


```python
a = np.array([[1,2],[3,4]])
for index, x in np.ndenumerate(a):
    print (index, x)
```

# create your own ufunc

```python
def area_rect(x,y):
    return x*y

#create an ufunc with 2 input 1 output
funcarea = np.frompyfunc(area_rect, 2, 1) #<-- ( function, 2 : input, 1 : output)

arr1 =[1,2,3,4]
arr2 =[2,3,4,5]
funcarea(arr1, arr2)
```

```python
np.frompyfunc(area_rect, 2, 1)(arr1, arr2) #<-- 縮寫
```

# @np.vectorize 《- input all kind of types and convert it to np.array

```python
# method 2
@np.vectorize
def area_rect2(x,y):
    return x*y

arr1 =[1,2,3,4]
arr2 =[2,3,4,5]
area_rect2(arr1, arr2)
```

# filer/ Search with no.where

```python
arr = np.array(np.arange(1,6))
arr
```

```python
arr [np.where(arr>=3)] # show value
```

```python
# if element >=3 -0.1, else +0.1
np.where(arr >=3, arr -0.1 ,arr +0.1) #<--- number >3 就 -0.1， else + 0.1
```

```python
np.where(arr>= 3, 'pass', 'Fail') #如果大過3， pass. 細過就fail
```
