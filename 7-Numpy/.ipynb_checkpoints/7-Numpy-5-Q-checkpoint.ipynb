## 47. Using broadcasting to perform element-wise operations
<pre>Transpose arr2 and add arr1
Given Input:
arr1 = np.array([[1, 2], [3, 4], [5,6]])
arr2 = np.array([2, 3, 4])

Desired Output:  
array([[ 3,  4],
       [ 6,  7],
       [ 9, 10]])
</pre>

```python
arr1 = np.array([[1, 2], [3, 4], [5,6]])
arr2 = np.array([2, 3, 4])

result = arr1 + arr2[:, None]
result
```

```python
arr1 = np.array([[1, 2], [3, 4], [5,6]])
arr2 = np.array([2, 3, 4])

result = arr1 + arr2[..., None]
result
```

## 48. Using broadcasting to perform element-wise operations
<pre>Transpose arr2 and multiply arr1
Given Input:
arr1 = np.array([[1, 2], [3, 4], [5,6]])
arr2 = np.array([2, 3, 4])

Desired Output:
array([[ 2,  4],
       [ 9, 12],
       [20, 24]])    

```python
arr1 = np.array([[1, 2], [3, 4], [5,6]])
arr2 = np.array([2, 3, 4])

result = arr1 * arr2[:, None]
result
```

## 49. Solve linear algebra
<pre>
Find the value of a and b
1a + 2b = 5
3a + 4b = 6

Desired Output:
[-4.   4.5]
</pre>

```python

A = np.array([[1, 2],
              [3, 4]])

B = np.array([5, 6])

solution = np.linalg.solve(A, B)
solution
```

## 50. Solve linear algebra
<pre>Find the value of x y, and z
3x + 4y + 2z = 1
2x + 5y + 3z = 2
x + 2y + z = 3

Desired Output:
[ -5.  12. -16.]</pre>

```python
x = np.array([[3,4,2],
             [2,5,3],
             [1,2,1]])
y = np.array([1,2,3])

solution = np.linalg.solve(x,y)
solution
```
