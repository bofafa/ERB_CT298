# Use for-loop to slice and print first 5 STR

```python
check_length = 4

for i in range(len(load_text)):
    if i < 5:
        print(load_text[i : i + check_length ])
```

# Find a STR pattern and print index number

```python
pattern = 'AGATC'

for i in range(len(load_text)):
    if load_text[i : i + len(pattern) ] == pattern :
        print(f'Found {pattern} on index {i}' )
```

# Find STR pattern and return repeat times

```python
pattern2 = 'TATC'
str_counter = 0

for i in range(len(load_text)):
    previous_found_index = np.nan
    
    
    if (load_text[i : i + len(pattern2) ] == pattern2) and (previous_found_index + len(pattern2) == i):
        previous_time_found_index = i
        str_counter += 1   # add str_counter 
        
    elif load_text[i : i + len(pattern2) ] == pattern2 :
        previous_time_found_index = i
        str_counter += 1
        
    print(f'Found {pattern2} repeat {str_counter}' )
```
