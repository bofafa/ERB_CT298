# *args <- arguments

```python
def sum_num( *args):
    total_args = 0
    for arg in args:
        total_args += arg
    return total_args

print(sum_num(1))
print(sum_num(1,2,3))
print(sum_num(1,2,3,2))

```

```python
def multiply_num(*args) -> float:
    """Multiply numbers"""
    total_nums = 1
    for num in args:
        total_nums *= num
    return total_nums

print(multiply_num(1, 2))          
print(multiply_num(1, 2, 3))       
print(multiply_num(2, 3, 4)) 
```

```python
def calculate (*args, **kwargs):
    """param: 
            *args: vaious of number
            **kwargs: method = 'add' for add up
            method = 'multiply' for mytiply all num"""
    for item in kwargs.items():
        if item[0] == 'method':
            method = item[1]
            if method == 'add':
                value = 0
                for arg in args:
                    value += arg

            elif method == 'multiply':
                value = 1
                for arg in args:
                    value *= arg

            else:
                raise Exception('Please input method="add" or mthod="multiply"')
                                
            return value
        else: 
            raise Exception('Please input method="add" or mthod="multiply"')

calculate(2,2,3, method = 'add')
```

# 搵座標 write a funtion  include 2 tuples Y & X locationand calucate the distance of X and Y


```python
data = [(1,1), (2,2)]  # 俾data 個形，for input

def cal_distance(date: list) -> float:
    """Calculate the distance of two points
       param: list of two points
    """
    x1, y1 = date[0][0], date[0][1]   # 用傳入嘅參數 date
    x2, y2 = date[1][0], date[1][1]

    d = (abs(x1 - x2)**2 + abs(y1 - y2)**2) ** 0.5
    
    return d

cal_distance(data)
```

```python
def finMinDis(points) -> dict:
    """find the minimum distance of 3 points"""

    result = dict()

    for i in range(len(points)):    
      for j in range(i, 3):
       if i != j:
           result[str(f'{i}-{j}')] = cal_distance([points[i], points[j]])

    return min(result, key = result.#get)


print(finMinDis([(4,5), (-1,-1), (1,-1),(0, 0.5)]))
```

# decarator @ example

```python
def outter(func):
    def inner():
        x = func()
        return x**2
    return inner

@outter
def pass_num():
    return 3

pass_num()
```

```python
def square_num(func):
    def inner(num):
        x = func(num)
        return x**2
    return inner
    
@square_num
def pass_num(num):
    return(num)

pass_num(3)
```

```python
import time

#define the decorator function
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func (*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} took {end_time - start_time: .2f} seconds to excute")
        return result
    return wrapper

# Apply the decorator to a function
@timer_decorator
def slow_function(sec):
    time.sleep(sec)

#call the decoRATED FUNCTION
slow_function(2)
```

```python
def power_num(func):
    def inner(*args, **kwargs):
        nums = func(*args, **kwargs)        # (1,2,3,4)

        # step 2: 逐個 square（LOOP 在這裡）
        squared = [n ** 2 for n in nums]    # [1,4,9,16]

        return squared
    return inner


def sum_num(func):
    def inner(*args, **kwargs):
        squared_list = func(*args, **kwargs)   # [1,4,9,16]

        total = 0
        for n in squared_list:                 # 再 SUM
            total += n

        return total
    return inner


@sum_num      # 外層 → 第二步（SUM）
@power_num    # 內層 → 第一步（SQUARE）
def pass_num(*nums):
    return nums     # simply return 原本的 arguments


print(pass_num(1, 2, 3, 4))   # 👉 30
```

```python
foo = 0 # < - x
def outer ():
    foo = 5 #< -x
    def middle():
        foo = 10 # <- O
        def inner():
            nonlocal foo # here
            foo += 1
            print(foo) #11
        inner()
    middle()
outer()
```

```python
foo = 0 # < - x


print('global foo address: ', id(foo), ', value =', foo)
def outer ():
    foo = 5 #< -x
    print('outer nonlocal foo address: ', id(foo), ', value =', foo)
    def middle():
        foo = 10 # <- O
        print('middle nonlocal foo address: ', id(foo), ', value =', foo)
        def inner():
            #print('foo address before global assignment: ', id(foo), ', value =', foo)
            nonlocal foo # here
            foo += 1
            print('inner golbal foo address: ', id(foo), ', value =', foo)
            print(foo) #11
        inner()
    middle()
outer()
```

```python
def scope_test():
    def do_local():
        spam='local spam'
        
    def do_nonlocal():
        nonlocal spam
        spam = 'nonlocal spam'
        
    def do_global():
        global spam
        spam = 'global spma'
    
    spam = 'test spam'
    
    do_local()
    print('after local assignment:', spam)
    do_nonlocal()
    print('after nonlocal assignment:', spam)
    do_global()
    print('after global assignment:', spam)

scope_test()
print('In global assignment:', spam)
```

# Regular Expression RE

```python
import requests

def gen_report():
    print("Hong Kong Weather Report")
    print(forecast['updateTime'])
    print("[Forecast]")
    print(forecast['forecastDesc'])
    print("[Outlook]")
    print(forecast['outlook'])  

try:
    language = 'tc'    # 'en'
    response = requests.get(f"https://data.weather.gov.hk/weatherAPI/opendata/weather.php?dataType=flw&lang={language}")
    forecast = response.json()
    gen_report()
except Exception as e:
    print("Error", e)
    
    
```

```python
import requests
from argparse import ArgumentParser, ArgumentDefaultsHelpFormatter

parser = ArgumentParser(prog='python hkweather.py',
    description='Hong Kong Weather Forecast香港天氣預報',
    formatter_class=ArgumentDefaultsHelpFormatter)

parser.add_argument("-l", "--language", help="Language (en, tc)")

args = vars(parser.parse_args())

def gen_report():
    print("Hong Kong Weather Report")
    print(forecast['updateTime'])
    print("[Forecast]")
    print(forecast['forecastDesc'])
    print("[Outlook]")
    print(forecast['outlook'])  

try:
    language = args["language"]    
    response = requests.get(f"https://data.weather.gov.hk/weatherAPI/opendata/weather.php?dataType=flw&lang={language}")
    forecast = response.json()
    gen_report()
except:
    print("usage: python hkweather.py [-h] [-l LANGUAGE]")
    
# usage 
# python hkweather.py -l tc
# python hkweather.py -l en
```
