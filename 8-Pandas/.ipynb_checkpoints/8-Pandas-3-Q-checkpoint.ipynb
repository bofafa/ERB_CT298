## 20. Replace values where the condition is False
<pre>Let's assume a species is large if its SepalLength is bigger
than mean. Create a new column 'LargeSpecies' and fill with 
Boolean to identify if the species belong to LargeSpecies.

You may create a column to fill with False first.
$> df.loc[df.index, 'LargeSpecies'] = False
</pre>

```python
df['SepalLength'].mean()
```

```python
#一開始俾晒 0 落新colnum先， 全False 
df.loc[df.index, 'LargeSpecies'] = 0
df['LargeSpecies'] = df['LargeSpecies'].astype('bool')
df.sample(2)
```

```python
# 
df['LargeSpecies'] = df['LargeSpecies'].where(
    df['SepalLength']<df['SepalLength'].mean(), True)
df.sample(5)
```

Series.where() & Series.mask()

```python
p = pd.Series([1,2,3,4,5])
p
```

```python
# replace values where the condition is False 
p.where(p<=3, 'Where P<=3 is false')
#pandas where, find false
```

```python
# replace values where the condition is True
p.mask(p<=3, 'Where P<=3 is true')
#panda marks, find  True
```

## 21. Replace values where the condition is True
<pre>df.mask() is the opposite of df.where()
df.mask(cond, other) will replace value with other if cond is True.

Let's create another column
df.loc[df.index, 'SmallSpecies'] = 0
df['SmallSpecies'] = df['SmallSpecies'].astype('bool')
Replace the Boolean to True if 'SepalLength' smaller and equal
to mean.
</pre>

```python
df.loc[df.index, 'SmallSpecies'] = 0
df['SmallSpecies'] = df['SmallSpecies'].astype('bool')
df
```

```python
df['SampleSize'] = np.where(
    df['SepalLength'] > df.SepalLength.mean(),
    'Large',
    'Small'
)
df
```

## 25. Map dict or function on a column

```python
df['SampleSize'] = df['SampleSize'].map({'SMALL': 'Small Iris', 
                                         'LARGE': 'Large Iris'})
```

```python
df.sample(5)
```

```python
df['SampleSize'] = df['SampleSize'].apply(lambda x: 'Large Iris' if x == 'Large' else 'Small Iris')
df.sample(5)
```
