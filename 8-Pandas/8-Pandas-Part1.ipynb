# Concat DataFrame

```python
import yfinance as yf
df_gold_jan = yf.download("GC=F", start="2023-01-01", end="2023-02-01"
                         ).droplevel('Ticker', axis=1)
df_gold_jan.head(1)
```

```python
df_gold_feb = yf.download("GC=F", start="2023-02-01", end="2023-03-01"
                         ).droplevel('Ticker', axis=1)
df_gold_feb.head(1)
```

```python
df_gold = pd.concat([df_gold_jan, df_gold_feb])
df_gold
```

```python
df_gold.shape
```

```python
df_buysell = pd.DataFrame({'BuySell' : np.zeros(df_gold.shape[0])})    
# Create a DF having same rows
df_buysell 
```

```python
df_gold = pd.concat([df_gold.reset_index(), df_buysell], axis=1).set_index('Date')    
# concat 2 df with same index
df_gold
```

```python
df_gold.info()
```

# Merge DataFrame

```python
df1 = pd.DataFrame({'Name': ['Alex', 'Bob', 'Chris'], 'ID': [1, 2, 3]})
df1
```

```python
import matplotlib.pyplot as plt
fig, (ax1,ax2) = plt.subplots(2,1, sharex=True, figsize=(15, 10),
                             gridspec_kw={'height_ratios':[3,1]})
ax1.plot( df_hsi.index, df_hsi['Close'], df_hsi['MA10'])
ax1.set_ylabel('Share Price')
ax1.legend(['Close', 'MA10'])
ax1.set_title('Closing Price')
ax2.bar( df_hsi.index, df_hsi['Volume'] )
ax2.set_ylabel('Daily volume')
ax2.set_title('Volume')
ax1.grid()
ax2.grid()
fig.suptitle('Hang Seng Index')
plt.show()
```

# Interacting with Database

```python
import sqlite3
```

SQL = Structural Query Language

```python
query = """
CREATE TABLE sales(
firstname VARCHAR(20), 
lastname VARCHAR(20),
age INTEGER,
sales REAL        
);
"""
```

```python
con = sqlite3.connect('sales.sqlite')
con.execute(query)   # you can create a table once
```

```python
con.commit()
```

```python
data = [('David', 'Chan', 35, 68000.50),
        ('Edgar', 'Doe', 45, 79000.60),
        ('Fion', 'Elle', 30, 88650.30)]
stmt = "INSERT INTO sales VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
```

```python
con.commit()
```

```python
cursor = con.execute('select * from sales')
rows = cursor.fetchall()
rows
```

```python
cursor.description
```

```python
df_sales = pd.DataFrame(rows, 
        columns=[x[0] for x in cursor.description])
df_sales
```
