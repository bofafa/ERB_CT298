# Gross Domestic Product - Table 310-30001 : Percentage change of Gross Domestic Product (GDP) and selected major expenditure components in real terms
`info`: https://data.gov.hk/en-data/dataset/hk-censtatd-tablechart-310-30001<BR>
`json`: https://www.censtatd.gov.hk/api/get.php?id=310-31001&lang=en&full_series=1<BR>
`docs`: https://www.censtatd.gov.hk/datagovhk/WT_data_dict_en.pdf<BR>

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://www.censtatd.gov.hk/api/get.php?id=310-31001&lang=en&full_series=1"
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
response = requests.get(url, headers=headers)
data = response.json()
```

```python
data['dataSet']
```

```python
df = pd.DataFrame(data['dataSet'])
df
```

```python
df.loc[(df.freq=='Y')&(df.sv=='CUR')&(df.svDesc=='Year-on-year % change')]
```

```python
df.loc[(df.freq=='Y')&(df.sv=='CUR')&(df.svDesc=='HK$ million')]
```

```python
df_chart = df.loc[(df.freq=='Y')&(df.sv=='CUR')&(df.svDesc=='HK$ million')]
fig, ax = plt.subplots(1,1, figsize=(12, 6))
ax.plot(df_chart.period, df_chart.figure)
plt.xticks(rotation=90, ha='center')
plt.title('Hong Kong GDP')
plt.grid(True)
plt.show()
```
