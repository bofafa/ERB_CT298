# Mini Project - Gold & Gold Stock

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
# Data Frame
df = pd.read_csv("gold_goldstock.csv")
df["Date"] = pd.to_datetime(df['Date'], dayfirst=True)  
df = df.set_index('Date')
df
```

```python
print('Correlation between GoldSpot and Zijin: ', df['GoldSpot'].corr(df['Zijin']) )
print('Correlation between GoldSpot and Zhaojin: ',df['GoldSpot'].corr(df['Zhaojin']) )
```

```python
#add constant to predictor variables
X = sm.add_constant(df['GoldSpot'])

#fit linear regression model
model = sm.OLS(df['Zijin'], X).fit()

#view model summary
summary = model.summary()
print(summary)
```

```python
#add constant to predictor variables
X = sm.add_constant(df['GoldSpot'])

#fit linear regression model
model = sm.OLS(df['Zhaojin'], X).fit()

#view model summary
summary = model.summary()
print(summary)
```

```python
sns.pairplot(df, x_vars=["Zijin", "Zhaojin"], y_vars=["GoldSpot"],
             height=10, aspect=.8, kind="reg");
```

# Gold ETF

```python
 import yfinance as yf
```

```python
daystart, dayend = '2024-01-01', '2025-12-03'

gold_future = yf.download('GC=F', start=daystart, end=dayend,
                          auto_adjust=True, progress=False).droplevel('Ticker', axis=1)

gold_etf = yf.download('7299.HK', start=daystart, end=dayend,
                       auto_adjust=True, progress=False).droplevel('Ticker', axis=1)

gold_future['NormPrx'] = gold_future['Close'] / gold_future['Close'].iloc[0]
gold_etf['NormPrx'] = gold_etf['Close'] / gold_etf['Close'].iloc[0]
```

```python
gold_future['Close'].corr(gold_etf['Close'])
```

```python
import matplotlib.pyplot as plt

plt.plot(gold_future.index, gold_future['NormPrx'], label='Future')
plt.plot(gold_etf.index, gold_etf['NormPrx'], label='ETF')

plt.legend()
plt.grid()
plt.show()
```
