```python
def collatz_1(x):
    original = x
    counter = 0
    while x != 1:
        if x % 2 == 0:
            x =x/2
        else:
            x = x*3+1
        counter += 1
    return original, x, counter
```

```python
collatz_1(90)
```

```python
 for n in ranint:
    print(collatz_1(n))
```
