```python
# Import required libraries
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# Configuration
TICKER = "GOOG"
START_DATE = "2024-05-22"
END_DATE = "2024-06-18"
INTERVAL = "1h"  # 1 hour candlesticks
VOLUME_WIDTH_PCT = 0.30  # Volume profile width as % of chart
VALUE_AREA_PCT = 0.70  # Value area contains 70% of volume

# Download data
print(f"Downloading {TICKER} data from {START_DATE} to {END_DATE}...")
df = yf.download(TICKER, start=START_DATE, end=END_DATE, interval=INTERVAL)

# Flatten MultiIndex columns if present
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

df = df.reset_index()

print(f"✓ Data shape: {df.shape}")
print(f"✓ Columns: {df.columns.tolist()}")
print(f"✓ Date range: {df['Datetime'].min()} to {df['Datetime'].max()}")
df.head()
```

```python
# Calculate technical indicators
# VWAP: Volume Weighted Average Price
df['VWAP'] = (df['Volume'] * (df['High'] + df['Low'] + df['Close']) / 3).cumsum() / df['Volume'].cumsum()

# Price range for volume profile
price_min = df['Low'].min()
price_max = df['High'].max()
price_bins = np.linspace(price_min, price_max, 50)

# Volume Profile: histogram of volume at each price level
volume_profile = []
for i in range(len(price_bins) - 1):
    mask = (df['Low'] <= price_bins[i + 1]) & (df['High'] >= price_bins[i])
    vol = df[mask]['Volume'].sum()
    volume_profile.append(vol)

volume_profile = np.array(volume_profile)

# Find POC (Point of Control): price with highest volume
max_vol_index = np.argmax(volume_profile)
poc_price = (price_bins[max_vol_index] + price_bins[max_vol_index + 1]) / 2

# Find Value Area: price levels containing 70% of total volume
sorted_indices = np.argsort(volume_profile)[::-1]
total_volume = volume_profile.sum()
cumulative_volume = 0
value_area_indices = []

for idx in sorted_indices:
    cumulative_volume += volume_profile[idx]
    value_area_indices.append(idx)
    if cumulative_volume >= VALUE_AREA_PCT * total_volume:
        break

value_area_prices = price_bins[value_area_indices]
va_top = value_area_prices.max()
va_bottom = value_area_prices.min()
current_price = df['Close'].iloc[-1]

print(f"✓ VWAP calculated")
print(f"✓ POC (Point of Control): ${poc_price:.2f}")
print(f"✓ Value Area: ${va_bottom:.2f} - ${va_top:.2f}")
print(f"✓ Current Price: ${current_price:.2f}")
```

```python
# Create chart
fig, ax1 = plt.subplots(1, 1, figsize=(16, 8))
ax2 = ax1.twiny()  # Secondary axis for volume profile

# === VOLUME PROFILE (background layer) ===
volume_scale = (len(df) * VOLUME_WIDTH_PCT) / volume_profile.max()
right_edge = len(df) * VOLUME_WIDTH_PCT

for i, vol in enumerate(volume_profile):
    price_level = (price_bins[i] + price_bins[i + 1]) / 2
    color = '#ef5350' if i in value_area_indices else '#26a69a'
    normalized_vol = vol * volume_scale
    
    ax2.barh(price_level, normalized_vol, left=right_edge - normalized_vol,
             height=(price_bins[1] - price_bins[0]) * 0.8,
             color=color, alpha=0.1, zorder=0)

# === CANDLESTICKS ===
for idx, row in df.iterrows():
    color = '#26a69a' if row['Close'] >= row['Open'] else '#ef5350'
    
    # Wick (high-low line)
    ax1.plot([idx, idx], [row['Low'], row['High']], 
             color=color, linewidth=1, zorder=2)
    
    # Body (open-close rectangle)
    height = abs(row['Close'] - row['Open'])
    bottom = min(row['Open'], row['Close'])
    ax1.add_patch(mpatches.Rectangle((idx - 0.3, bottom), 0.6, height,
                                      facecolor=color, edgecolor=color, zorder=3))

# === INDICATORS ===
# VWAP line
ax1.plot(df.index, df['VWAP'], color='#2196F3', linewidth=1.5,
         linestyle='--', alpha=0.7, label='VWAP', zorder=4)

# POC line and label
ax1.axhline(y=poc_price, color='purple', linestyle='-',
           linewidth=2, alpha=0.7, label='POC', zorder=4)
ax1.text(len(df) + 0.5, poc_price, f'${poc_price:.2f}',
         verticalalignment='center', fontsize=9,
         bbox=dict(boxstyle='round', facecolor='purple', alpha=0.7, edgecolor='none'),
         color='white', fontweight='bold', zorder=5)

# Value Area lines
ax1.axhline(y=va_top, color='#2196F3', linestyle='--',
           linewidth=1, alpha=0.5, label='Value Area Top', zorder=4)
ax1.axhline(y=va_bottom, color='#2196F3', linestyle='--',
           linewidth=1, alpha=0.5, label='Value Area Bottom', zorder=4)

# Current price indicator
ax1.axhline(y=current_price, color='black', linestyle='-',
           linewidth=1.5, alpha=0.8, zorder=5)
ax1.text(len(df) + 0.5, current_price, f'{current_price:.2f}',
         verticalalignment='center', fontsize=10,
         bbox=dict(boxstyle='round', facecolor='#2196F3', alpha=0.8, edgecolor='none'),
         color='white', fontweight='bold', zorder=5)

# === CHART FORMATTING ===
ax1.set_xlim(-1, len(df))
ax1.set_ylim(price_min * 0.99, price_max * 1.01)
ax1.set_ylabel('Price ($)', fontsize=12, fontweight='bold')
ax1.set_xlabel('')
ax1.set_title(f'{TICKER}, {INTERVAL} Candles\nVWAP | Volume Profile | POC',
              fontsize=12, loc='left', pad=20, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle='-', linewidth=0.5, zorder=1)
ax1.legend(loc='upper left', fontsize=9)

# X-axis date labels
date_col = 'Datetime' if 'Datetime' in df.columns else 'Date'
date_labels = df[date_col].dt.strftime('%d. %b').unique()[::5]
date_positions = [i for i in range(0, len(df), len(df) // len(date_labels) + 1)][:len(date_labels)]
ax1.set_xticks(date_positions)
ax1.set_xticklabels(date_labels, rotation=0, ha='center')

# Hide volume profile axis
ax2.set_xlim(0, len(df) * VOLUME_WIDTH_PCT)
ax2.set_ylim(price_min * 0.99, price_max * 1.01)
ax2.axis('off')

# Styling
fig.patch.set_facecolor('white')
ax1.set_facecolor('white')
ax1.tick_params(colors='black', labelsize=9)
for spine in ax1.spines.values():
    spine.set_color('black')
ax1.xaxis.label.set_color('black')
ax1.yaxis.label.set_color('black')
ax1.title.set_color('black')

plt.tight_layout()
plt.show()

# Summary
print(f"\n{'='*50}")
print(f"CHART SUMMARY")
print(f"{'='*50}")
print(f"Ticker:           {TICKER}")
print(f"Current Price:    ${current_price:.2f}")
print(f"POC:              ${poc_price:.2f}")
print(f"Value Area High:  ${va_top:.2f}")
print(f"Value Area Low:   ${va_bottom:.2f}")
print(f"VWAP:             ${df['VWAP'].iloc[-1]:.2f}")
print(f"{'='*50}")
```
