```python
class Staff:
    def __init__(self, name, sid):  #__init__ 初始化 method, staff 必須有嘅element，self = staff
        self.name = name  #self 隱色參數 implicit argument，直接知道self 係staff
        self.sid = sid
        # name, sid 顯色參數，要話明有乜用
        
    def verify(self):
        return f"{self.name} ID: {self.sid}"

```

```python
import math

class DataMan:
    def __init__(self, data: list):  
        self.data = data              

    def addData(self):                
        return sum(self.data)

    def mulData(self):
        return math.prod(self.data)

data = DataMan([1, 2, 3, 4])
print(data.addData())
print(data.mulData())
```

```python
class StockCheck:
    def __init__(self, stockdata: list):
        self.stockdata = stockdata

    def calReturn(self):
        return (self.stockdata[-1] - self.stockdata[0])/ self.stockdata[0]

    def DailyReturn(self):
        dailyRtn = []  
        for i in range(len(self.stockdata)):
            today = self.stockdata[i]
            if i == len(self.stockdata)-1: break     #因為有突，所以- 1 
            dailyRtn.append(self.stockdata[i+1]/self.stockdata[i]-1)  # (110-100)/ 100 = 110/100 -1 /// (110/100) - (100/100)
        return dailyRtn



stock = StockCheck([26095,25760, 25953, 26085, 25919])
print ("Return:", stock.calReturn())
print(f'return: {stock.calReturn()*100:.4f}%') #.2f 小數點後2 個float


print("Daily Returns:", stock.DailyReturn())
```

```python
class Staff:
    def __init__(self, name):
        self.name = name

    def verify(self):
        print (f'Registered staff')

class SalesStaff(Staff):
    def display(self):
        print ('Sales staff name:', self.name)
```

```python
class Staff:  #<--- parent
    name = ''
    def display(self):
        print(f'Registered staff')

class SalesStaff(Staff):   # <-- child
    def display(self):
        print ('Sales staff name:', self.name) # display child
        super().display() #<--- display 用返parent
```

```python
sales_1 = SalesStaff()

sales_1.name = 'Gary Wong'

sales_1.display()
```

# multple inheritance

```python
class P1:
    def __init__(self, name):
        self.name = name
    def p1method(self):
        print ('p1 method')

class P2:
    def __init__(self, sid):
        self.sid = sid

class SalesStaff(P1, P2):
    def __init__(self, name, sid, email):
        P1.__init__(self, name)
        P2.__init__(self, sid)
        self.email = email
        
    def display(self):
        print("Sales staff name:", self.name, "; Sale staff ID:", self.sid, "; Email: ", self.email  )

sale_01 = SalesStaff('Alice', 'A12345', 'alice@abc.com')
sale_01.display()
sale_01.p1method()
```

# composion

```python
class Employee:    
    def __init__ (self, name, address):
        self.name = name
        print ('Composite - Employee object created...')
        self.address = Address(address)

    def emp_info(self):
        print (f'Employee name: {self.name}')
        print('composit - emp_info() method executed...')
        self.address.print_add()

class Address:
    def __init__(self, address):
        self.addr = address
        print ('Composite - Address object created...')

    def print_add(self):
        print (f'Employee name: {self.addr}')
        print ('Composite - Address print_add() method also executed...')
    
```

```python
staff_01 = Employee("john Lee","28 mody Road, Tsim sha Tsui, kowloon")
```

```python
staff_01.emp_info()
```

```python
class DataProcess:
    def __init__ (self, data):
        self.data = data
    def mean(self):
        return sum(self.data) / len (self.data)
    def std(self):
        m = self.mean()
        return (sum((x-m)**2 for x in self.data) / len(self.data))**0.5
    def zscore(self):
        m = self.mean()
        std = self.std()
        return [(x-m)/std for x in self.data]
    def minmax(self):
        min_val = min(self.data)
        max_val = max (self.data)
        return [(x- min_val)/ (max_val - min_val) for x in self.data]
```

```python
data1 = DataProcess([1,2,3,4])
print('mean = ', data1.mean())
print('Standard Deviation = ', data1.std())
print('Z-score = ', data1.zscore())
print('MinMax', data1.minmax())
```
