# Time Series with Range Slider

```python
import yfinance as yf

nvda = yf.download("NVDA", start="2013-01-01", end="2023-01-01"
                  ).droplevel('Ticker', axis=1)
```

```python
fig = px.line(nvda, x=nvda.index, y=nvda['Close'], 
              title="NVidia Share Price (US$)"
            ).update_layout(yaxis_title="Share Price (close) US$")
fig.update_xaxes(rangeslider_visible=True,
            rangeselector=dict(
            buttons=list([
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=2, label="2y", step="year", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(step="all")
        ])))
fig.show()
```

```python
df = pd.DataFrame({'code':['kyg','abc']})
df.loc[df.code.str.startswith('kyg')]
```

```python
stockList = pd.read_excel('ListOfSecurities.xlsx', skiprows=2) #2
stockList = stockList.loc[
    (stockList['Sub-Category']=='Equity Securities (Main Board)') &
    (stockList['Category']=='Equity')]
stockList  # 2363 rows  18 columns
```

```python
def changeTicker(code):
    return (4-len(str(int(code))))*'0'+str(int(code))+'.HK'
changeTicker('00001')
```

```python
hkMainBoard = [changeTicker(code) for code in stockList['Stock Code']]
```

```python
len(hkMainBoard)
```

```python
%%time
from tqdm import tqdm
#tickers = ["NVDA","TSM"]
def downloadStock(ticker):
    df = yf.download(ticker, start="2025-01-01", end="2025-10-29",
                      auto_adjust=True, progress=False
                     ).droplevel('Ticker', axis=1)
    df['Ticker'] = ticker
    return df[['Ticker','Close','High','Low','Open','Volume']]
df_hkstocks = pd.concat(downloadStock(t) for t in tqdm(hkMainBoard[:100]))
```

```python
df_hkstocks.to_csv('df_hkstocks.csv')
```

```python
def featEng(ticker):  
    try:
        df = df_hkstocks.loc[df_hkstocks.Ticker==ticker]
        df.loc[df.index, ['Return']] = df.Close.pct_change(1)
        df = df.dropna()
        return df
    except:
        pass
df_feat = pd.concat(featEng(t) for t in hkMainBoard) 
df_feat
```

# Moving Avarage and Exponential Moving Average

```python
df_g = yf.download("GOOG", start="2022-01-01", end="2023-01-01"
                  ).droplevel('Ticker', axis=1)
df_g
```

```python
df_g['MA20'] = df_g["Close"].rolling(20).mean()
df_g['EMA20'] = df_g["Close"].ewm(span=20, adjust=False).mean()
df_g
```

```python
df_g = df_g.dropna()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1, shared_xaxes=True, 
                    vertical_spacing=0.01, row_heights=[0.7, 0.3])

# fig.add_candlestick(x=df_g.index, open=df_g['Open'], high=df_g['High'],
#                 low=df_g['Low'], close=df_g['Close'],
#                 increasing_line_color= 'blue', 
#                     decreasing_line_color= 'red',
#                 name='Close',row=1,col=1 )
fig.add_trace(go.Candlestick(x=df_g.index, open=df_g['Open'], high=df_g['High'],
                low=df_g['Low'], close=df_g['Close'],
                increasing_line_color= 'blue', 
                    decreasing_line_color= 'red',
                name='Close' ),  row=1,col=1)

# fig.add_scatter(x=df_g.index, y=df_g['MA20'], name='MA20',row=1,col=1)               
# fig.add_scatter(x=df_g.index, y=df_g['EMA20'], name='EMA20',row=1,col=1) 
fig.add_trace(go.Scatter(x=df_g.index, y=df_g['MA20'], name='MA20'),  row=1,col=1)
fig.add_trace(go.Scatter(x=df_g.index, y=df_g['EMA20'], name='EMA20'),  row=1,col=1)

fig.add_trace(go.Bar(x=df_g.index, y=df_g.Volume,
                     showlegend=True,name='Volume',
                    marker=dict(color='rgb(125,125,222)')),row=2,col=1)

fig.update_layout(title="Alphabet Inc. (GOOG) Share Price (Close) US$",
                    yaxis_title="Close Price US$",
                    xaxis_rangeslider_visible=False,
                    width=1000, height=800)
fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]), #hide weekends
        dict(values=["2022-07-01", "2022-10-01"])  # hide Special date
    ]
)
fig.show()
```

# Fitting multiple lines and retrieving the model parameters

```python
import plotly.express as px

df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", facet_col="smoker", color="sex", trendline="ols")
fig.show()

results = px.get_trendline_results(fig)
print(results)

results.query("sex == 'Male' and smoker == 'Yes'").px_fit_results.iloc[0].summary()
```
